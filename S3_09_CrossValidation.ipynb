Validación Cruzada de Retención (Hold-out)

Validación Cruzada k-fold

Aunque se utiliza la validación cruzada, se debe reservar un conjunto de prueba independiente para la evaluación final. En este caso, ya no es necesario un conjunto de validación adicional.


k-fold Cross Validation (Validación cruzada k-fold)

En este enfoque básico:

El conjunto de entrenamiento se divide en k subconjuntos (folds) más pequeños.

Para cada uno de los k folds, se realiza el siguiente procedimiento:

Se entrena un modelo usando k − 1 folds como datos de entrenamiento.

El modelo resultante se valida en el fold restante (que actúa como conjunto de prueba para esa iteración) para calcular una medida de desempeño (por ejemplo, precisión).

La medida de desempeño final de la validación cruzada k-fold es el promedio de las puntuaciones obtenidas en cada iteración.



Método Hold-out (Retención)

Se divide el conjunto de datos inicial en dos conjuntos: entrenamiento y prueba.

El conjunto de entrenamiento se usa para entrenar el modelo.

El conjunto de prueba se usa para estimar su rendimiento de generalización.

Una variación consiste en dividir el conjunto de entrenamiento en dos partes:

Conjunto de entrenamiento: para ajustar distintos modelos.

Conjunto de validación: para ajustar y comparar diferentes configuraciones de parámetros, mejorar el rendimiento en datos no vistos y seleccionar el mejor modelo.

Este proceso se llama selección de modelo (model selection).

El objetivo es encontrar los valores óptimos de los parámetros de ajuste (hiperparámetros).



k-fold Cross Validation (detalle)

Se divide aleatoriamente el conjunto de entrenamiento en k folds, sin reemplazo.

Se entrena el modelo en k – 1 folds.

El fold restante se usa para evaluar el rendimiento.

Este procedimiento se repite k veces, cambiando el fold de evaluación.

Resultados finales:

Se obtienen k modelos y k estimaciones de rendimiento.

Se calcula el promedio del rendimiento para obtener una estimación más robusta y menos sensible a cómo se partió el conjunto de entrenamiento (en comparación con el método hold-out).

La validación cruzada k-fold se usa para ajustar el modelo y encontrar los hiperparámetros óptimos que proporcionan un buen rendimiento de generalización.

Una vez que se encuentran los valores óptimos de hiperparámetros, se puede:

Reentrenar el modelo con todo el conjunto de entrenamiento

Evaluar el rendimiento final con el conjunto de prueba independiente

Esto tiene sentido porque, al entrenar con más datos, generalmente se obtiene un modelo más preciso y robusto.

Un valor común de k es 10.

Para conjuntos de entrenamiento relativamente pequeños, conviene usar un valor de k más alto.



Validación Cruzada k-fold Estratificada (Stratified k-fold)

Es una variación de la validación cruzada k-fold.

Produce mejores estimaciones de sesgo y varianza, especialmente en casos donde las clases no están equilibradas (proporciones desiguales).
